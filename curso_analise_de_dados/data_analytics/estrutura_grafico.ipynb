```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# cria a figura
plt.figure(figsize= (5, 5))
# criar o plot
plt.plot([15, 20, 99], [13, 90, 10],
        color = 'blue',
        linewidth = 1.5,
        linestyle = '--',
        )

plt.plot([99, 56, 1], [45, 23, 99])

# labels = rótulos
# rótulo do eixo x
plt.xlabel('Eixo x')
# rótulo do eixo y
plt.ylabel('Eixo y')
# título da figura
plt.title('Minha figura')

plt.grid(alpha = 0.5)
```

```python
fig, ax = plt.subplots(2, 2, figsize = (6, 6))

ax[0, 0].plot([1, 2, 3], [5, 5, 9])
ax[0, 0].set(xlabel = 'Eixo X', ylabel = 'Eixo Y')


ax[0, 1].plot([1, 2, 3], [5, 5, 9])

ax[1, 0].plot([1, 2, 3], [5, 5, 9])

ax[1, 1].plot([1, 2, 3], [5, 5, 9])

plt.tight_layout(pad = 4)
```
