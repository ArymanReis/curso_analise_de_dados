```python
import numpy as np
import pandas as pd
import seaborn as sns

# Definir tema
sns.set_theme(style='darkgrid')
```

```python
# Dados
Base_Dados = sns.load_dataset('tips')

# Varificar
Base_Dados.head()
```

```python
# Renomeando base de dados
Base_Dados_Pt = Base_Dados.rename( columns= {
  'total_bill': 'Total Conta',
  'tip': 'Gorjeta',
  'sex': 'Sexo',
  'smoker': 'Fumante',
  'day': 'Dia Da Semana',
  'time': 'Período',
  'size': 'Pessoas'
})
```

```python
Base_Dados_Pt
```

```python
# Gráfico relplot
sns.relplot( x='Total Conta', y='Gorjeta', data=Base_Dados_Pt)
```

```python
# Gráfico relplot
sns.relplot( x='Total Conta', y='Gorjeta', data=Base_Dados_Pt, hue='Sexo')
```

```python
# Gráfico linhas com dois eixos
sns.relplot( x='Total Conta', y='Gorjeta', data=Base_Dados_Pt, kind='line')
```

```python
# Gráfico de linha com 2 eixos
sns.relplot( x='Total Conta', y='Gorjeta', data=Base_Dados_Pt, kind='line', hue='Fumante')
```

```python
# Gráfico
sns.histplot( x='Total Conta', data=Base_Dados_Pt)
```

```python
# Gráfico
sns.histplot( x='Total Conta', data=Base_Dados_Pt, hue='Fumante')
```

```python
# Gráfico
sns.barplot( x='Sexo', y='Total Conta', data=Base_Dados_Pt, hue='Fumante')
```

```python
sns.pairplot(Base_Dados_Pt)
```

```python
sns.pairplot(Base_Dados_Pt, hue='Sexo')
```

```python
sns.pairplot(Base_Dados_Pt, kind='kde')
```

```python
sns.boxplot(data=Base_Dados_Pt, x='Dia Da Semana', y='Total Conta')
```

```python
sns.boxplot(data=Base_Dados_Pt, x='Dia Da Semana', y='Total Conta', hue='Sexo')
```
